```python
import pandas as pd
df = pd.read_csv('Iris.csv')
df
```

```python
samples = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
samples
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(samples)
```

```python
labels = model.predict(samples)
labels
```

```python
new_samples = [[5.7,4.4,1.5,0.4],
               [6.5,3.0,5.5,1.8],
               [5.8,2.7,5.1,1.9]]
new_labels = model.predict(new_samples)
new_labels
```

```python
import matplotlib.pyplot as plt
xs = samples[:,0]
ys = samples[:,2]
plt.scatter(xs,ys,c=labels)
plt.show()
```

```python
model.inertia_
```

```python
species = df["Species"].values
sp_lb_df = pd.DataFrame({'labels':labels,'species':species})
sp_lb_df
```

```python
ct = pd.crosstab(sp_lb_df.labels,sp_lb_df.species)
ct
```
